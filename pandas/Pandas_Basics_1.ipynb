## describe()

- ```If we are interested to know the basic statistics of our data we will use .describe() method```<br>
- ```It tells us only the statistics of numeric data```

````python
TitanicData.describe()
````

- ```It gives us only the statistics of numeric columns```<br>
- **count:** ```count means the total number of observations```<br>
- **mean:** ```It tells the average of each series```<br>
- **Standard Deviation:** ```It measures the dispersion of a dataset relative to its mean & it is the square root of variance```<br>
- **min & max:**```It tells the maximum & minimum value of each series```<br>
- **25%:**```It is the first quartile(lower quartile)```<br>
- **50%:**```It is the second quartile(median)```<br>
- **75%:**```It is the third quartile(upper quartile)```<br>

## loc function():

- ```It is label based data selecting method which means that we have to pass the name of the row or column which we want to select.```<br>

### Select single row:

````python
TitanicData.loc[0]
````

### Select single column:

````python
TitanicData.loc[:,'Age'] #similar to TitanicData['Age']
````

### Select multiple rows:

````python
TitanicData.loc[[0,1,2]]
````

### Select row range:

````python
TitanicData.loc[0:6] #last index also includes
````

### Selecting multiple columns:


````python
TitanicData.loc[:,['Age','Fare']]
````

### Selecting alternate columns:

````python
TitanicData.loc[:,'Age':'Fare':1]
````

### Using conditions:

````python
TitanicData.loc[TitanicData['Age']>29] 
````

## iloc function():

```It is an indexed-based selecting method which means that we have to pass an integer index in the method to select a specific row/column. This method does not include the last element of the range passed in it unlike loc().```

````python
singleRow = TitanicData.iloc[0] 
multiRow = TitanicData.iloc[0:10]
singleColumn = TitanicData.iloc[:,0] 
multiColumn = TitanicData.iloc[:,[0,3]]
alternateRows = TitanicData.iloc[0:4:1]
alternateColumns = TitanicData.iloc[:,0:4:1]
print(singleRow,multiRow,singleColumn,multiColumn,alternateRows,alternateColumns)
````

## Create new columns derived from existing columns:

````python
airQuality=pd.read_csv('https://raw.githubusercontent.com/pandas-dev/pandas/main/doc/data/air_quality_no2.csv')
airQuality
````

````python
airQuality['newDATA']= airQuality['station_paris']*100
airQuality
````

## Changing columns name:

````python
airQuality.rename(
    columns={
        'datetime':'DateTime',
        'station_antwerp':'Station_Antwerp',
        'station_paris':'Station_Paris',
        'station_london':'Station_London',
        'newDATA':'NewData'
    }
)
````

````python
airQuality.rename(columns=str.lower)
````

## Aggregate function:

````python
TitanicData.agg(
    {
        "Age":["min", "max", "median", "skew"],
        "Fare": ["min", "max", "median", "mean"],
    }
)
airQuality.agg(
    {
        "station_antwerp":["min","max","median","skew"],
        "station_paris":["min","max","median","skew"],
        "station_london":["min","max","median","skew"],
    }
)
````

## Concat:

```python
frames = [ df1,df2]
display(pd.concat(frames))
display(pd.concat(frames,axis=1))# concatenate horizontally
result = pd.concat(frames,keys=['X','Y']) # WE CREATE TABLE CONSIST OF CHUNKS X and Y
display(result)
result.loc["Y"] # now we access data from dataframe on the basis of chunk
```

## joins:

```python
result = pd.concat(frames,join='inner',axis=0)
result
```

## Pivot

```python
df3 = pd.read_csv("https://raw.githubusercontent.com/vishwajeetsinghrana8/pythonforml/main/Sales_Funnel_CRM.csv")
df3
```

### we need to find company and product wise liscenses:

```python
df4 = df3[['Company','Product','Licenses','Status']]
df4.pivot(index='Company',columns='Product',values='Licenses')
```

### Multi index pivot 

```python
df5 = df4.pivot(index=['Company','Product'],columns='Status',values='Licenses')
df5
```

## Stacked:

```python
stack = df5.stack()
stack
```

```python
stack.unstack(0)
```

```python
stack.unstack(1)
```

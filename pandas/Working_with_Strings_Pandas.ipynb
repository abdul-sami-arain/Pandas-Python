## Splitting and replacing strings

```python
data['Age group'].str.split('-')
```

__Elements in the split lists can be accessed using get or [] notation:__

```python
data['Age group'].str.split('-').get(1)
```

```python
data['Age group'].str.split('-')[1]
```

```python
data['Age group'].str.split('-',expand=True) # it makes easy to get answer in dataFrame format
```

__rsplit is similar to split except it works in the reverse direction__

### replace()
in below example i replace all commas with "and"

```python
data['Which data subject area are you most interested in working with? (Select up to three)'].str.replace(", "," and ",case=False,regex=True)
```

```python
pattern = r'[A-z]+'
def repl(m):
    return m.group()[::-1]
data['What do you hope to get out of the workshop?'].str.replace(pattern,repl,regex=True)
#group is a function of regex python
```
